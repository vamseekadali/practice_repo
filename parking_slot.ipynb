```python
# these variables are setting vehicle limit  for every floor
floor1 = 5
floor2 = 5
floor3 = 5
while True:
# this block checks whether the parking slots are complete or not
    if floor1 == 0 and floor2 == 0 and floor3 == 0:
        print("\nparking is full")
        break
    vehicle = input("\nenter vehicle type car or bike or truck or exit: ").lower()
    if vehicle == "exit":
        print("thank you! visit again.")
        break
    if vehicle != "car" and vehicle != "bike" and vehicle != "truck":
        print("invalid vehicle type")
        continue
    hours = int(input("enter parking hours: "))
    # initiate the rate based on vehicle type
    if vehicle == "bike":
        rate = 20
    elif vehicle == "car":
        rate = 50
    else:
        rate = 100
    base_amount = rate * hours

    assigned_floor = 0
    # this block checks the vehicle type and assigns the floor based on availability
    if vehicle == "truck":
        if floor1 > 0:
            assigned_floor = 1
            floor1 = floor1 - 1
        else:
            print("Parking Full")
            continue
    elif vehicle == "car":
        if floor1 > 0:
            assigned_floor = 1
            floor1 = floor1 - 1
        elif floor2 > 0:
            assigned_floor = 2
            floor2 = floor2 - 1
        else:
            print("Parking Full")
            continue
    elif vehicle == "bike":
        if floor1 > 0:
            assigned_floor = 1
            floor1 = floor1 - 1
        elif floor2 > 0:
            assigned_floor = 2
            floor2 = floor2 - 1
        elif floor3 > 0:
            assigned_floor = 3
            floor3 = floor3 - 1
        else:
            print("parking full")
            continue
    # this block calculates the extra charges based on parking hours
    if hours <= 2:
        extra = 0
    elif hours <= 5:
        extra = base_amount * 10 / 100
    elif hours <= 10:
        extra = base_amount * 20 / 100
    else:
        extra = base_amount * 30 / 100
    total = base_amount + extra
    # this block calculates the discount based on vehicle type
    if vehicle == "bike":
        discount = total * 5 / 100
    elif vehicle == "car":
        discount = total * 10 / 100
    else:
        discount = total * 15 / 100
    final_bill = total - discount
    print("\n---- receipt ----")
    print("vehicle type   :", vehicle)
    print("assigned floor :", assigned_floor)
    print("parking hours  :", hours)
    print("base amount    :", base_amount)
    print("extra charges  :", extra)
    print("discount       :", discount)
    print("final bill     :", final_bill)
    print("\nRemaining Slots")
    print("floor 1 :", floor1)
    print("floor 2 :", floor2)
    print("floor 3 :", floor3)
```
