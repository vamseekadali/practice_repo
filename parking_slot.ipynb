```python
# these variables are setting vehicle limit  for every floor
floor1 = 5
floor2 = 5
floor3 = 5
while True:
# this block checks whether the parking slots are complete or not
    if floor1 == 0 and floor2 == 0 and floor3 == 0:
        print("\nParking is Full")
        break
    vehicle = input("\nEnter Vehicle Type (Car/Bike/Truck/Exit): ").lower()
    if vehicle == "exit":
        print("Thank You! Visit Again.")
        break
    if vehicle != "car" and vehicle != "bike" and vehicle != "truck":
        print("Invalid Vehicle Type")
        continue
    hours = int(input("Enter Parking Hours: "))
    # Base Price
    if vehicle == "bike":
        rate = 20
    elif vehicle == "car":
        rate = 50
    else:
        rate = 100
    base_amount = rate * hours

    assigned_floor = 0
    # Floor Allocation
    if vehicle == "truck":
        if floor1 > 0:
            assigned_floor = 1
            floor1 = floor1 - 1
        else:
            print("Parking Full")
            continue
    elif vehicle == "car":
        if floor1 > 0:
            assigned_floor = 1
            floor1 = floor1 - 1
        elif floor2 > 0:
            assigned_floor = 2
            floor2 = floor2 - 1
        else:
            print("Parking Full")
            continue
    elif vehicle == "bike":
        if floor1 > 0:
            assigned_floor = 1
            floor1 = floor1 - 1
        elif floor2 > 0:
            assigned_floor = 2
            floor2 = floor2 - 1
        elif floor3 > 0:
            assigned_floor = 3
            floor3 = floor3 - 1
        else:
            print("Parking Full")
            continue
    # Extra Charges
    if hours <= 2:
        extra = 0
    elif hours <= 5:
        extra = base_amount * 10 / 100
    elif hours <= 10:
        extra = base_amount * 20 / 100
    else:
        extra = base_amount * 30 / 100
    total = base_amount + extra
    # Discount
    if vehicle == "bike":
        discount = total * 5 / 100
    elif vehicle == "car":
        discount = total * 10 / 100
    else:
        discount = total * 15 / 100
    final_bill = total - discount
    print("\n========== PARKING RECEIPT ==========")
    print("Vehicle Type   :", vehicle.capitalize())
    print("Assigned Floor :", assigned_floor)
    print("Parking Hours  :", hours)
    print("Base Amount    :", base_amount)
    print("Extra Charges  :", extra)
    print("Discount       :", discount)
    print("Final Bill     :", final_bill)
    print("\nRemaining Slots")
    print("Floor 1 :", floor1)
    print("Floor 2 :", floor2)
    print("Floor 3 :", floor3)
```
